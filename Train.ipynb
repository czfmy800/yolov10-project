```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolov10s.yaml")  # build a new model from YAML
model = YOLO("yolov10s.pt")  # load a pretrained model (recommended for training)
model = YOLO("yolov10s.yaml").load("yolov10s.pt")  # build from YAML and transfer weights

# Train the model
results = model.train(data="TrafficSign.yaml", epochs=100, imgsz=640)
```
